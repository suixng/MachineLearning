```python
def load_data(path, transpose=True):
    data = sio.loadmat(path)
    y = data.get('y') #(5000,1)
    y = y.reshape(y.shape[0]) #make it back to column vector
    
    X = data.get('X') #(5000,400)
    
    if transpose:
        # for this dataset, you need a transpose to get the orientation right
        X = np.array([im.reshape((20, 20)).T for im in X])
        
        # and I flat the image again to preserve the vector presentation
        X = np.array([im.reshape(400) for im in X])
        
    return X, y
```

## train 1 model(训练一维模型)

```python
def cost(theta, X, y):
    # cost fn is -1(theta) for you to minimize
    return np.mean(-y * np.log(sigmoid(X @ theta)) - (1 -y) * np.log(1 - sigmoid(X @ theta)))
```

```python
def regularized_cost(theta, X, y, l=1):
    theta_j1_to_n = theta[1:]
    regularized_term = (1 / (2 * len(X))) * np.power(theta_j1_to_n, 2).sum()
    
    return cost(theta, X, y) + regularized_term
```

```python
def regularized_gradient(theta, X, y, l=1):
    theta_j1_to_n = theta[1:]
    regularized_theta = (1 / len(X)) * theta_j1_to_n
    
    #by doing this, no offset is on theta_0
    regularized_term = np.concatenate([np.array([0]), regularized_theta])
    
    return gradient(theta, X, y) + regularized_term
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def gradient(theta, X, y):
    return (1 / len(X)) * X.T @ (sigmoid(X @ theta) - y)
```

```python
def logistic_regression(X, y, l=1):
    # init theta
    theta = np.zeros(X.shape[1])
    
    #train it
    res = opt.minimize(fun=regularized_cost,
                      x0=theta,
                      args=(X, y, 1),
                      method='TNC',
                      jac=regularized_gradient,
                      options={'disp':True})
    # get trained parameters
    final_theta = res.x
    
    return final_theta
```

```python
def predict(x, theta):
    prob = sigmoid(x @ theta)
    return (prob >= 0.5).astype(int)
```

```python
t0 = logistic_regression(X, y[0])
# print(t0)
```

```python
print(t0.shape)
y_pred = predict(X, t0)
print('Accuracy={}'.format(np.mean(y[0] == y_pred)))
```

## train K model(训练K维模型）

```python
k_theta  = np.array([logistic_regression(X, y[k]) for k in range(10)])
print(k_theta.shape)
```

## 进行预测

- think about the shape of k_theta, now you are making X * /theta.T
> (5000,401) * (10,401).T = (5000,10)
- after that, you run sigmoid to get probabilities and for each row, you find highest prob as the answer

```python
prob_matrix = sigmoid(X @ k_theta.T)
```

```python
np.set_printoptions(suppress=True)
prob_matrix
```

```python
y_pred = np.argmax(prob_matrix, axis=1) #返回沿轴axis最大值的索引，axis=1代表行
```
